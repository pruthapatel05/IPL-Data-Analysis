```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
#Getting the frequency of most man of the match awards
man_of_match = df['player_of_match'].value_counts()
```

```python
print(man_of_match)
```

```python
man_of_match.head()
```

```python
#making a bar-plot for the top 5 players with most man of the match awards
plt.figure(figsize = (8,5))
plt.bar(list(man_of_match[0:5].keys()),man_of_match[0:5], color = 'g')
plt.show()
```

```python
#Extracting the records where a team won batting first
batting_first = df[df['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
plt.hist(batting_first['win_by_runs'])
```

```python
batting_first['winner'].value_counts()
```

```python
#Making a bar-plot for top 3 teams with most wins after batting first
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),batting_first['winner'].value_counts()[0:3])
plt.show()
```

```python
plt.pie(list(batting_first['winner'].value_counts()), labels = batting_first['winner'].value_counts().keys(),autopct = "%0.1f%%")
plt.show()
```

```python
#extracting those records where a team has won after batting second
batting_second = df[df['win_by_wickets']!=0]
```

```python
batting_second.head()
```

```python
batting_second['winner'].value_counts()
```

```python
#Making a histogram for frequency of wins w.r.t number of wickets
plt.hist(batting_second['win_by_wickets'], bins = 30)
plt.show()
```

```python
#Making a bar plot for top-3 teams with most wins after batting second
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),batting_second['winner'].value_counts()[0:3])
plt.show()
```

```python
plt.pie(batting_second['winner'].value_counts(), labels = batting_second['winner'].value_counts().keys())
plt.show()
```
